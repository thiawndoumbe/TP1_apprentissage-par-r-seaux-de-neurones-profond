```python
from deeplib.datasets import load_mnist, train_valid_loaders
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torchvision.transforms import ToTensor
import numpy as np
import torch
import random
from deeplib.visualization import plot_images
import torch
import time
import matplotlib.pyplot as plt

mnist_data, test_data = load_mnist()
mnist_data.transform = ToTensor()
test_data.transform = ToTensor()
```

```python
class NeuralNet(nn.Module):
    def __init__(self, input_dim, output_dim, hidden_layers):
        super().__init__()
        self.layers = nn.ModuleList([nn.Linear(input_dim, input_dim) for _ in range(hidden_layers)])
        self.final_layer = nn.Linear(input_dim, output_dim)

    def forward(self, x):
        x = x.flatten(1)
        for layer in self.layers:
            x = F.relu(layer(x))
        x = self.final_layer(x)
        return x

class NeuralNetSoftmax(nn.Module):
    def __init__(self, input_dim, hidden_units, output_dim):
        super().__init__()
        self.layers = nn.ModuleList([nn.Linear(input_dim, input_dim) for _ in range(hidden_units)])
        self.final_layer = nn.Linear(input_dim, output_dim)

    def forward(self, x):
        x = x.flatten(1)
        for layer in self.layers:
            x = F.relu(layer(x))
        x = self.final_layer(x)
        x = F.softmax(x, dim=1)
        return x
```

```python
def compute_accuracy(true_labels, predicted_labels):
    correct = sum(1 for true, pred in zip(true_labels, predicted_labels) if true == pred)
    return correct / len(true_labels)

def validate_model(model, loss_function, data_loader, mse=False, verbose=False):
    true_labels, predicted_labels, image_predictions = [], [], []
    validation_losses = []

    model.eval()
    with torch.no_grad():
        for inputs, targets in data_loader:
            inputs, targets = inputs.cuda(), targets.cuda()
            outputs = model(inputs)
            predictions = outputs.max(dim=1)[1]

            if verbose:
                for img, label in zip(inputs, predictions):
                    image_predictions.append({'img': img.cpu().numpy().squeeze(0), 'label': label.cpu().numpy()})

            target_labels = F.one_hot(targets, num_classes=10).float() if mse else targets
            loss = loss_function(outputs, target_labels)
            validation_losses.append(loss.item())

            true_labels.extend(targets.cpu().numpy().tolist())
            predicted_labels.extend(predictions.cpu().numpy().tolist())

    return compute_accuracy(true_labels, predicted_labels), np.mean(validation_losses), image_predictions
```

```python
batch_size, epochs, lr = 128, 18, 0.1
def train_model(model, mse=False):
    model.cuda()
    loss_function = nn.MSELoss() if mse else nn.CrossEntropyLoss()
    optimizer = optim.SGD(model.parameters(), lr=lr, momentum=0.2)
    train_loader, valid_loader = train_valid_loaders(mnist_data, batch_size)

    train_acc_list, test_acc_list, train_loss_list, test_loss_list = [], [], [], []
    start_time = time.time()

    model.train()
    for epoch in range(1, epochs + 1):
        epoch_start = time.time()
        epoch_loss, true_labels, predicted_labels = [], [], []

        for data, labels in train_loader:
            data, labels = data.cuda(), labels.cuda()
            optimizer.zero_grad()
            outputs = model(data)
            labels_onehot = F.one_hot(labels, num_classes=10).float() if mse else labels
            loss = loss_function(outputs, labels_onehot)
            predictions = outputs.max(dim=1)[1]
            loss.backward()
            optimizer.step()

            epoch_loss.append(loss.item())
            true_labels.extend(labels.cpu().numpy().tolist())
            predicted_labels.extend(predictions.cpu().numpy().tolist())

        train_accuracy = compute_accuracy(true_labels, predicted_labels)
        val_accuracy, val_loss, _ = validate_model(model, loss_function, valid_loader, mse)

        train_acc_list.append(round((1 - train_accuracy) * 100, 1))
        test_acc_list.append(round((1 - val_accuracy) * 100, 1))
        test_loss_list.append(val_loss)
        train_loss_list.append(np.mean(epoch_loss))

        print(f'Epoch {epoch} - Train Acc: {train_accuracy*100:.2f} - Val Acc: {val_accuracy*100:.2f} - Loss: {np.mean(epoch_loss):.4f} - Time: {time.time() - epoch_start:.2f}s')

    print(f'Total Training Time: {time.time() - start_time:.2f}s')
    return train_acc_list, test_acc_list, test_loss_list, train_loss_list

def plot_training_results(train_accuracy, test_accuracy, filename):
    epochs = np.arange(1, len(train_accuracy) + 1)
    plt.figure(figsize=(12, 8))
    plt.plot(epochs, train_accuracy, label='Train Error')
    plt.plot(epochs, test_accuracy, label='Validation Error')
    plt.xticks(np.arange(0, len(train_accuracy) + 1, step=4))
    plt.xlabel('Epochs')
    plt.ylabel('Error Rate (%)')
    plt.title('Training and Validation Error Rate')
    plt.legend()
    plt.grid(True)
    plt.savefig(filename)
    plt.show()
input_dim, output_dim, hidden_layers = 28*28, 10, 4

model = NeuralNet(input_dim, output_dim, hidden_layers)
train_acc, test_acc, test_loss, train_loss = train_model(model)
plot_training_results(train_acc, test_acc, 'Training_Results')


```

**Test avec MSE**

```python
model = NeuralNet(input_dim, output_dim, hidden_layers)
train_acc, test_acc, test_loss, train_loss = train_model(model, mse=True)
plot_training_results(train_acc, test_acc, 'MSE_Training_Results')


softmax_model = NeuralNetSoftmax(input_dim, hidden_layers, output_dim)
train_acc, test_acc, test_loss, train_loss = train_model(softmax_model, mse=True)
plot_training_results(train_acc, test_acc, 'Softmax_Training_Results')
plot_training_results(train_loss, test_loss, 'loss_mse_result')
```
